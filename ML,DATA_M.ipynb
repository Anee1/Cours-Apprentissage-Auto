 #  Définition de la Data Mining

La Data Mining, également appelée exploration de données, est le processus d'analyse de grandes quantités de données pour découvrir des modèles, des tendances, des relations ou des informations cachées. En d'autres termes, c'est l'extraction de connaissances significatives à partir de données brutes, souvent stockées dans des bases de données ou d'autres formes de stockage de données.

##  Objectifs et applications

Les objectifs du Data Mining sont multiples et peuvent varier en fonction des domaines d'application. Voici quelques objectifs généraux :

Découverte de modèles et de tendances : Identifier des schémas récurrents, des relations et des structures cachées dans les données.

Prédiction et prévision : Faire des prédictions sur des événements futurs ou des valeurs inconnues en se basant sur des données historiques.

Segmentation de données : Regrouper les données en segments homogènes pour une meilleure compréhension des différents groupes.

Classification : Classer les données dans des catégories ou des classes prédéfinies en se basant sur leurs caractéristiques.

Détection d'anomalies : Identifier les données qui se comportent de manière inhabituelle ou qui sont des exceptions par rapport au reste des données.

Recommandation : Fournir des recommandations personnalisées aux utilisateurs en se basant sur leurs préférences et comportements passés.

Quant aux applications du Data Mining, elles sont vastes et touchent de nombreux domaines, notamment :

Marketing : Analyse de la clientèle, segmentation de marché, recommandation de produits, prévision des ventes.

Finance : Détection de fraudes, analyse de risques, prévision des tendances boursières, scoring de crédit.

Santé : Diagnostic médical, analyse des données patients, détection de maladies, prévision de l'évolution des épidémies.

Science et Recherche : Exploration de grandes bases de données scientifiques, modélisation de processus complexes, analyse de séquences génétiques.

Web et Médias Sociaux : Analyse des tendances sur les réseaux sociaux, recommandation de contenu, détection de spams

## Importance dans les domaines des affaires et de la recherche

Dans les Affaires :
Prise de Décision Éclairée : Les entreprises utilisent le Data Mining pour prendre des décisions stratégiques basées sur des analyses approfondies des données. Cela peut inclure des décisions en matière de marketing, de gestion des stocks, de planification financière, etc.

Optimisation des Processus : Le Data Mining permet d'identifier les inefficacités dans les processus opérationnels et de les optimiser pour améliorer l'efficacité et la rentabilité.

Segmentation de Marché : En analysant les données des clients, les entreprises peuvent identifier les segments de marché et adapter leurs produits ou services pour répondre aux besoins spécifiques de chaque segment.

Prévision des Tendances et des Comportements : Les modèles de Data Mining peuvent être utilisés pour prédire les tendances du marché, les comportements des clients, les demandes futures de produits, etc.

Détection de Fraude : Les algorithmes de Data Mining peuvent détecter les schémas frauduleux dans les transactions financières, les demandes de remboursement d'assurance, etc.

Dans la Recherche :
Analyse des Données Scientifiques : Les chercheurs utilisent le Data Mining pour explorer de vastes ensembles de données scientifiques, trouver des corrélations entre les variables, et découvrir de nouvelles informations.

Modélisation de Processus Complexes : Le Data Mining permet de modéliser et de simuler des processus complexes dans des domaines tels que la biologie, la chimie, la physique, etc.

Découverte de Nouvelles Connaissances : En analysant de grandes quantités de données, les chercheurs peuvent découvrir de nouveaux modèles, relations, et tendances qui pourraient ne pas être évidents à première vue.

Prédiction et Diagnostic : Les techniques de Data Mining sont utilisées pour prédire les résultats de recherches, diagnostiquer des maladies à partir de données médicales, etc.

Optimisation de la Recherche : En identifiant les domaines à fort potentiel de recherche et en ciblant les efforts de recherche, le Data Mining peut contribuer à une utilisation plus efficace des ressources de recherche.

## Nettoyage de la Base de Données
Nous allons supprimer la colonne "A_id" qui ne fournit pas d'informations pertinentes. Étant donné que toutes les valeurs manquantes se trouvent sur une seule ligne, nous allons supprimer cette ligne. En outre, comme observé, la colonne "Acidity" est une variable numérique ; nous allons donc la transformer en variable numérique. Enfin, nous allons recoder la variable "Quality" :  "good" en 1 et "bad" en 0

```python
def clean_data(df):
    
    df = df.drop(columns=['A_id'])
    
    df = df.dropna()
    
    df = df.astype({'Acidity': 'float64'})
    
    df['Quality'] = df['Quality'].map({'good': 1, 'bad': 0})
    
    return df
```

```python
data_0 = clean_data(data)
```

En examinant le graphe de corrélation, nous constatons que nos variables ne présentent pas de corrélation linéaire significative. Cela signifie qu'il n'y a pas de relation directe et proportionnelle entre les variables étudiées. Ces résultats suggèrent que les variations dans une variable ne sont pas nécessairement accompagnées de variations dans une autre de manière constante.

Cette absence de corrélation linéaire peut avoir plusieurs implications pour notre analyse. Par exemple, cela peut signifier  les relations entre elles sont plus complexes.

La classification est une tâche fondamentale de l'apprentissage automatique supervisé où l'objectif est de prédire une variable de sortie (ou cible) discrète ou catégorique en fonction des caractéristiques (ou des variables d'entrée) associées à chaque observation. En d'autres termes, la classification consiste à attribuer des étiquettes ou des classes à des données en fonction de certaines caractéristiques

## Voting Classifier

```python
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
```

```python
model_1 = SGDClassifier(random_state=0)
model_2 = DecisionTreeClassifier(random_state=0)
model_3 = KNeighborsClassifier(n_neighbors=2)

model_4 = VotingClassifier([('SGD', model_1),
                            ('Tree', model_2),
                            ('KNN', model_3)],
                          voting='hard')

for model in (model_1, model_2, model_3, model_4):
    model.fit(X_train, y_train)
    print(model.__class__.__name__, model.score(X_test, y_test))
```
